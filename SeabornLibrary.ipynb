# Seaborn Library

```
## How to plot a histpgram in Seaborn.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df= pd.read_csv("ai_supplychain.csv")

## basic histogram approach.
sns.histplot(data=df, x="AI_Adoption_Rate (%)", bins=15, color= "steelblue")
plt.title("Ditribution of AI Adoption Rate", fontsize=12, fontweight="bold")
plt.xlabel("AI Adoption Rate (%)")
plt.ylabel("Frequency")
plt.show()

## Plotting Histogram with NAs Dropped

sns.histplot(data=df.dropna(subset=["AI_Adoption_Rate (%)"]),
             x="AI_Adoption_Rate (%)", bins=15, color="green")
plt.title("AI Adoption Rate (NAs Dropped)")
plt.show()


## Setting color for Histogram.

sns.histplot(data=df,x="Productivity_Index (0-100)", color="orange", bins=20)
plt.title("Distribution of Productivity Index")
plt.show()

## The Type of plot that would allow the Comparison of two Constinous features are scaterplot or regression plot.
sns.regplot(data=df, x="AI_Adoption_Rate (%)", y="Productivity_Index (0-100)",
            scatter_kws={"color": "violet"}, line_kws={"color": "black"})
plt.title("Relationship Between AI Adoption and Productivity")
plt.show()

## A Correloation plot. (Heatmap)
corr= df[["AI_Adoption_Rate (%)", "AI_Tech_Used", "Productivity_Index (0-100)"]].corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Between AI Usage and Productivity")
plt.show()

## Change figure size of the plots

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x="AI_Tech_Used", color="skyblue", bins=10, edgecolor="black", alpha=0.7)
plt.title("Distribution of AI Technologies Used", fontsize=14, fontweight='bold')
plt.xlabel("AI Technologies Used", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)  # rotate labels for clarity
plt.tight_layout()
plt.show()

```
